```python
plt.figure(figsize=(5,5))
x = [1,2,3,4,5]

y = [10,20,60,40,50]

k = [45,78,26,50,35]

plt.plot(x,y)

plt.title("Simple Plot")
plt.xlabel("Time (hr)")
plt.ylabel("Position (km)")

plt.plot(k, 'y--')

plt.legend(labels=('Values of y','Values of k'), loc = 'upper right')

plt.xticks(ticks = [0,1,2,3,4,5], labels = ['0','x1','x2','x3','x4','x5'])

plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6,6), sharex=True, sharey=True)

ax[0].plot(y,'go')
ax[1].plot(k)

ax[0].set_xticks(ticks = [0,1,2,3,4,5])
ax[1].set_xticks(ticks = [0,1,2,3,4,5])

ax[0].set_xticklabels(labels=['0','x1', 'x2', 'x3', 'x4','x5']);
ax[1].set_xticklabels(labels=['0','x1', 'x2', 'x3', 'x4','x5']);
```

```python
import pandas as pd
import numpy as np
```

```python
data_BM = pd.read_csv('C:\\Users\\Owner\\Desktop\\Big-Mart-Sales-master\\Train.csv')
data_BM = data_BM.dropna(how='any')
data_BM.head()
```

```python
price_by_item = data_BM.groupby('Item_Type').Item_MRP.mean()
price_by_item
```

```python
price_by_item = data_BM.groupby('Item_Type').Item_MRP.mean()

x = price_by_item.index.tolist()
y = price_by_item.values.tolist()

plt.figure(figsize=(20,5))

plt.title('Mean price for each item type')

plt.xlabel('Item Type')
plt.ylabel('Mean Price')

plt.xticks(labels=x, ticks=np.arange(len(x)))

plt.plot(x, y)
```

```python
sales_by_outlet_size = data_BM.groupby('Outlet_Size').Item_Outlet_Sales.mean()

#sales_by_outlet_size.sort_values(inplace=False)

x = sales_by_outlet_size.index.tolist()
y = sales_by_outlet_size.values.tolist()

plt.xlabel('Outlet Size')
plt.ylabel('Sales')

plt.title('Mean sales for each outlet type')

plt.xticks(labels=x, ticks=np.arange(len(x)))

plt.bar(x, y, color=['red', 'orange', 'teal'])
```

```python
plt.title('Item price distribution')

plt.xlabel('Item_MRP')
plt.ylabel('Frequency')

plt.hist(data_BM['Item_MRP'],bins=30, color='pink')
```

```python
data = data_BM[['Item_Outlet_Sales']]

red_diamond = dict(markerfacecolor='r', marker='D')

plt.title('Item Sales distribution')

plt.boxplot(data.values, labels=['Item Sales'], flierprops=red_diamond);
```
